```python
# ==========================================================
# STEP 12: Remove Duplicate Rows
# ==========================================================

print("Duplicate Rows Before Removal:", df.duplicated().sum())

# Remove duplicate records
df = df.drop_duplicates()

print("Duplicate Rows After Removal:", df.duplicated().sum())
print("Dataset Shape After Removing Duplicates:", df.shape)


# ==========================================================
# STEP 13: Check Missing Values Again
# ==========================================================

print("\nMissing Values After Duplicate Removal:")
print(df.isnull().sum())

# Fill any remaining missing numeric values with median
df.fillna(df.median(numeric_only=True), inplace=True)

print("\nMissing Values After Filling:")
print(df.isnull().sum())


# ==========================================================
# STEP 14: Validate Important Columns
# ==========================================================

# Keep only valid age values
df = df[(df["age"] >= 1) & (df["age"] <= 120)]

# Blood pressure should be greater than 0
df = df[df["trestbps"] > 0]

# Cholesterol should be greater than 0
df = df[df["chol"] > 0]

# Maximum heart rate should be greater than 0
df = df[df["thalach"] > 0]

print("\nDataset Shape After Validation:", df.shape)


# ==========================================================
# STEP 15: Detect Outliers Using IQR Method
# ==========================================================

numeric_columns = ["age", "trestbps", "chol", "thalach", "oldpeak"]

print("\nOutlier Count:")

for col in numeric_columns:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_limit) | (df[col] > upper_limit)]

    print(f"{col}: {len(outliers)} outliers")


# ==========================================================
# STEP 16: Cap Outliers (Winsorization)
# ==========================================================

for col in numeric_columns:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < lower_limit, lower_limit, df[col])
    df[col] = np.where(df[col] > upper_limit, upper_limit, df[col])

print("\nOutliers have been capped successfully.")


# ==========================================================
# STEP 17: Final Dataset Information
# ==========================================================

print("\nFinal Dataset Shape:", df.shape)

print("\nData Types:")
print(df.dtypes)

print("\nStatistical Summary:")
print(df.describe())


# ==========================================================
# STEP 18: Save Cleaned Dataset
# ==========================================================

df.to_csv("heart_cleaned.csv", index=False)

print("\n✅ Cleaned dataset saved successfully as 'heart_cleaned.csv'")


# ==========================================================
# STEP 19: Final Data Quality Report
# ==========================================================

quality_report = pd.DataFrame({
    "Data Type": df.dtypes,
    "Missing Values": df.isnull().sum(),
    "Unique Values": df.nunique()
})

print("\nData Quality Report:")
print(quality_report)


# ==========================================================
# STEP 20: Final Verification
# ==========================================================

print("\n" + "="*50)
print("FINAL DATASET SUMMARY")
print("="*50)
print(f"Rows               : {df.shape[0]}")
print(f"Columns            : {df.shape[1]}")
print(f"Missing Values     : {df.isnull().sum().sum()}")
print(f"Duplicate Rows     : {df.duplicated().sum()}")
print("="*50)
print("✅ Dataset is now clean and ready for Exploratory Data Analysis (Phase 5).")
```

```python

# =========================
# PHASE 6: STATISTICAL ANALYSIS
# =========================

import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import ttest_ind, chi2_contingency

print("\n==============================")
print("1. BASIC STATISTICAL SUMMARY")
print("==============================")
print(df.describe())

print("\n==============================")
print("2. MEDIAN VALUES")
print("==============================")
print(df.median(numeric_only=True))

print("\n==============================")
print("3. MODE VALUES (MOST COMMON)")
print("==============================")
print(df.mode().iloc[0])

print("\n==============================")
print("4. VARIANCE (DATA SPREAD)")
print("==============================")
print(df.var(numeric_only=True))

print("\n==============================")
print("5. STANDARD DEVIATION")
print("==============================")
print(df.std(numeric_only=True))

print("\n==============================")
print("6. CORRELATION WITH TARGET")
print("==============================")
print(df.corr(numeric_only=True)['target'].sort_values(ascending=False))

# =========================
# T-TEST (AGE vs HEART DISEASE)
# =========================
print("\n==============================")
print("7. T-TEST: AGE vs TARGET")
print("==============================")

disease = df[df['target'] == 1]['age']
no_disease = df[df['target'] == 0]['age']

t_stat, p_value = ttest_ind(disease, no_disease)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("👉 RESULT: Statistically SIGNIFICANT difference in age between groups")
else:
    print("👉 RESULT: NOT statistically significant")

# =========================
# CHI-SQUARE TEST (Chest Pain vs Disease)
# =========================
print("\n==============================")
print("8. CHI-SQUARE TEST: CP vs TARGET")
print("==============================")

table = pd.crosstab(df['cp'], df['target'])

chi2, p, dof, expected = chi2_contingency(table)

print("Chi-square Value:", chi2)
print("P-Value:", p)

if p < 0.05:
    print("👉 RESULT: Chest pain is SIGNIFICANTLY associated with heart disease")
else:
    print("👉 RESULT: No strong association")

# =========================
# CONFIDENCE INTERVAL (AGE)
# =========================
print("\n==============================")
print("9. 95% CONFIDENCE INTERVAL (AGE)")
print("==============================")

data = df['age']

mean = np.mean(data)
ci = stats.norm.interval(0.95, loc=mean, scale=stats.sem(data))

print("Mean Age:", mean)
print("95% CI:", ci)

print("\n==============================")
print("PHASE 6 COMPLETED SUCCESSFULLY")
print("==============================")
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

print("\n==============================")
print("PHASE 8: ADVANCED ANALYTICS (FIXED)")
print("==============================")

# =========================
# 1. CLEAN DATA FIRST (IMPORTANT FIX)
# =========================

print("\n------------------------------")
print("1. DATA CLEANING FOR ML")
print("------------------------------")

df_clean = df.copy()

# Convert categorical/object columns to numeric safely
for col in df_clean.columns:
    if df_clean[col].dtype == 'object' or str(df_clean[col].dtype) == 'category':
        df_clean[col] = pd.Categorical(df_clean[col]).codes

print("All categorical columns encoded.")

# =========================
# 2. FEATURE SEPARATION
# =========================

features = df_clean.drop('target', axis=1)

# Ensure everything is numeric
features = features.apply(pd.to_numeric, errors='coerce')

# Fill any missing values created during conversion
features = features.fillna(features.mean())

print("Feature matrix cleaned.")

# =========================
# 3. STANDARDIZATION
# =========================

scaler = StandardScaler()
scaled_data = scaler.fit_transform(features)

print("Data standardized successfully.")

# =========================
# 4. PCA
# =========================

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

df_pca = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])
df_pca['target'] = df_clean['target']

print("PCA variance ratio:", pca.explained_variance_ratio_)

plt.figure(figsize=(8,6))
sns.scatterplot(x='PC1', y='PC2', hue='target', data=df_pca)
plt.title("PCA - Heart Disease Structure")
plt.show()

# =========================
# 5. K-MEANS CLUSTERING
# =========================

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(scaled_data)

df_clean['cluster'] = clusters

print("\nCluster distribution:")
print(df_clean['cluster'].value_counts())

# =========================
# 6. CLUSTER VISUALIZATION
# =========================

plt.figure(figsize=(8,6))
sns.scatterplot(x=df_pca['PC1'], y=df_pca['PC2'], hue=df_clean['cluster'])
plt.title("Patient Clusters (PCA View)")
plt.show()

# =========================
# 7. RISK ANALYSIS
# =========================

cluster_risk = df_clean.groupby('cluster')['target'].mean()

print("\nCluster Risk Levels:")
print(cluster_risk)

print("\nClinical Interpretation:")
for i, r in cluster_risk.items():
    if r > 0.7:
        print(f"Cluster {i}: HIGH RISK")
    elif r > 0.4:
        print(f"Cluster {i}: MEDIUM RISK")
    else:
        print(f"Cluster {i}: LOW RISK")

print("\n==============================")
print("PHASE 8 COMPLETED SUCCESSFULLY")
print("==============================")
```

```python

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    roc_auc_score,
    recall_score
)

import matplotlib.pyplot as plt
import seaborn as sns

print("\n==============================")
print("PHASE 9: MACHINE LEARNING")
print("==============================")

# =========================
# 1. DATA PREPARATION
# =========================

df_ml = df.copy()

# Encode categorical columns safely
for col in df_ml.columns:
    if df_ml[col].dtype == 'object' or str(df_ml[col].dtype) == 'category':
        df_ml[col] = pd.Categorical(df_ml[col]).codes

X = df_ml.drop('target', axis=1)
y = df_ml['target']

X = X.fillna(X.mean())

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print("\nData prepared successfully.")

# =========================
# 2. CROSS VALIDATION SETUP
# =========================

def evaluate_model(model, name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else y_pred

    print("\n------------------------------")
    print(name)

    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Recall:", recall_score(y_test, y_pred))
    print("ROC-AUC:", roc_auc_score(y_test, y_proba))
    print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("\nClassification Report:\n", classification_report(y_test, y_pred))

    # Cross-validation score
    cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='accuracy')
    print("Cross Validation Accuracy:", cv_scores.mean())

    return model

# =========================
# 3. LOGISTIC REGRESSION (TUNED)
# =========================

log_model = GridSearchCV(
    LogisticRegression(max_iter=1000),
    param_grid={'C': [0.01, 0.1, 1, 10]},
    cv=5
)

log_model = evaluate_model(log_model, "Logistic Regression (Tuned)")

# =========================
# 4. RANDOM FOREST (TUNED)
# =========================

rf_model = GridSearchCV(
    RandomForestClassifier(),
    param_grid={
        'n_estimators': [50, 100, 200],
        'max_depth': [3, 5, 10]
    },
    cv=5
)

rf_model = evaluate_model(rf_model, "Random Forest (Tuned)")

# =========================
# 5. DECISION TREE (TUNED)
# =========================

dt_model = GridSearchCV(
    DecisionTreeClassifier(),
    param_grid={
        'max_depth': [3, 5, 10, None]
    },
    cv=5
)

dt_model = evaluate_model(dt_model, "Decision Tree (Tuned)")

# =========================
# 6. FEATURE IMPORTANCE (BEST MODEL)
# =========================

print("\n==============================")
print("FEATURE IMPORTANCE (Random Forest)")
print("==============================")

best_rf = rf_model.best_estimator_

importances = best_rf.feature_importances_

feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print(feature_importance)

plt.figure(figsize=(10,6))
sns.barplot(x='Importance', y='Feature', data=feature_importance)
plt.title("Feature Importance (Heart Disease Prediction)")
plt.show()

print("\n==============================")
print("PHASE 9 COMPLETED")
print("==============================")
```

```python

# =========================
# PHASE 13: PUBLIC HEALTH + DEPLOYMENT SYSTEM
# =========================

from IPython.display import Markdown, display

# =========================
# 1. RISK SCORING ENGINE
# =========================

def calculate_risk(row):
    score = 0

    # Age risk
    if row['age'] > 55:
        score += 2

    # Cholesterol risk
    if row['chol'] > 240:
        score += 2

    # Blood pressure risk
    if row['trestbps'] > 140:
        score += 2

    # Chest pain risk
    if row['cp'] != 0:
        score += 2

    # Low heart rate response
    if row['thalach'] < 120:
        score += 2

    # Exercise angina
    if row['exang'] == 1:
        score += 2

    return score


# Apply risk scoring
df['risk_score'] = df.apply(calculate_risk, axis=1)


# =========================
# 2. RISK CATEGORY SYSTEM
# =========================

def risk_category(score):
    if score >= 8:
        return "🔴 High Risk"
    elif score >= 4:
        return "🟠 Medium Risk"
    else:
        return "🟢 Low Risk"


df['risk_category'] = df['risk_score'].apply(risk_category)


# =========================
# 3. RISK DISTRIBUTION OUTPUT
# =========================

risk_distribution = df['risk_category'].value_counts()

display(Markdown("## 📊 Risk Category Distribution"))
display(risk_distribution.to_frame())


# =========================
# 4. MEDICAL INSIGHTS
# =========================

medical_insights = """
# 🫀 Medical Insights

- Age, cholesterol, and blood pressure are major risk drivers
- Chest pain is a strong early warning signal
- Low heart rate response indicates poor cardiac function
- Exercise-induced angina is a critical risk marker

---

# 🧠 Key Understanding:
CAD is a cumulative disease driven by multiple interacting risk factors.
"""

display(Markdown(medical_insights))


# =========================
# 5. HOSPITAL DEPLOYMENT STRATEGY
# =========================

hospital_strategy = """
# 🏥 Hospital Deployment Model

## 🔴 High Risk Patients
- Immediate cardiology referral
- ECG + angiography recommended
- ICU preparedness

## 🟠 Medium Risk Patients
- Lifestyle intervention
- Medication + follow-up
- Monthly monitoring

## 🟢 Low Risk Patients
- Preventive checkups
- Health education
- Annual screening
"""

display(Markdown(hospital_strategy))


# =========================
# 6. GOVERNMENT HEALTH POLICY INSIGHTS
# =========================

government_policy = """
# 🏛️ Government Policy Recommendations

- National heart screening program
- Free cholesterol and BP testing camps
- AI-based rural diagnostic systems
- Subsidized preventive healthcare programs

---

# 🌍 Public Health Insight:
Early detection reduces heart attack mortality significantly.
"""

display(Markdown(government_policy))


# =========================
# 7. BUSINESS IMPACT
# =========================

business_impact = """
# 💰 Business & Industry Impact

## Hospitals:
- Faster patient triage
- Reduced ICU overload

## Insurance:
- Risk-based premium calculation
- Lower claim losses

## Pharma:
- Targeted drug distribution
"""

display(Markdown(business_impact))


# =========================
# 8. FINAL SYSTEM SUMMARY
# =========================

final_summary = """
# 🚀 FINAL SYSTEM OUTPUT

✔ Risk scoring engine created
✔ Patient classification completed
✔ Hospital workflow model designed
✔ Government policy mapping completed
✔ Public health system framework built

🎯 FINAL RESULT:
An AI-powered preventive healthcare decision system for Coronary Artery Disease.
"""

display(Markdown(final_summary))
```
